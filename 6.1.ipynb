```python
#475
filename = input("Enter filename: ")

f = open(filename, "r")

count = 0   # to count lines printed

while True:
    line = f.readline()

    # If file ended
    if line == "":
        print("\n-- End of File --")
        break

    print(line, end="")
    count += 1

    # After printing 25 lines â†’ pause
    if count == 25:
        choice = input("\nEnter 'continue' or 'stop': ")

        if choice.lower() == "stop":
            print("File closed.")
            break
        elif choice.lower() == "continue":
            count = 0    # reset counter for next 25 lines
        else:
            print("Invalid input. Stopping program.")
            break

f.close()
```

```python
#477
# Basic program to extract 4-letter words starting and ending with same letter

with open("friends.txt", 'r') as file:
    text = file.read()

# Split the text into words
words = text.split()

# Extract four-letter words starting and ending with the same letter
four_letter_words = []
for word in words:
    if len(word) == 4 and word[0].lower() == word[-1].lower():
        four_letter_words.append(word)

if four_letter_words:
    print("List of four-letter words starting and ending with the same letter:")
    for word in four_letter_words:
        print(word)
else:
    print("No matching words found.")
```

```python
f = open("best1.txt")
l = f.readlines()
f.close()

result = ""
for i in l:
    k = i.split()
    for j in k:
        if j[0].lower() == "i":
            result += j[::-1] + " "
        else:
            result += j + " "
    result = result.strip() + "\n"

print(result.strip("\n"))
```

```python
# Open file to write user input
f = open("user_input.txt", "w")

while True:
    line = input("Enter Something (for quit enter END): ")
    if line == "END":
        break
    f.write(line + "\n")  # Write each line in a new line

f.close()

# Open file to read and display lines starting with capital letter
f = open("user_input.txt", "r")
print("The Line started with Capital Letters:")
for line in f:
    if line[0].isupper():
        print(line.strip())
f.close()
```

```python
# Open the mailbox file
f = open("box.txt", "r")

# Dictionary to store sender counts
counts = {}

for line in f:
    line = line.rstrip()
    # Process only lines that start with 'From ' (with a space)
    if line.startswith("From "):
        email = line.split()[1]  # second word is the sender email
        # Update the count in the dictionary
        counts[email] = counts.get(email, 0) + 1

f.close()

# Display the dictionary of counts
print(counts)

# Find the sender with the maximum count
max_sender = None
max_count = 0
for sender, count in counts.items():
    if count > max_count:
        max_sender = sender
        max_count = count

# Display the most prolific sender
print(max_sender, max_count)
```
