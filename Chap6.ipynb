```python
''' 
r+ : to read and write data into the file
the previous data in the file won't be deleted,the file pointer is placed at the begining of the file
r+ mode works only on existing file
'''
f=open("xyz.txt","r+")
f.write("xyz")
f.close()
```

```python
'''
w+:to write and read data,it will override existing data
will create new file if the file is not existing
'''
f=open("xyz.txt","w+")
f.write("India")
f.close()
```

```python
'''
a+: to append and read data from the file, it won't override existing 
will create new file if the file is not existing

'''
f=open("xyz.txt","a+")
f.write("Gujarat")
f.close()
```

```python
#Various property of file object
'''
1.name :> gives the name of file
2.mode :> gives the mode of file; ex:r,w,w+ etc...
3.closed :> returns True or False
4.readable() :> returns True or False
5.writable() :> returns True or False :>> will return false when file is in r mode
'''
```

```python
f=open("abc.txt","w")

print(f.name)
print(f.mode)
print(f.readable())
print(f.writable())
f.close()
print("file closed",f.closed)
```

Write a function display_oddLines() to display odd number lines from the text file. Consider the following lines for the file – 
friends.txt.
 Friends are crazy, Friends are naughty !
 Friends are honest, Friends are  best !
 Friends are like keygen, friends are like license key !
 We are nothing without friends, Life is not possible without friends !

```python
def odd_lines():
    f=open("friends.txt")
    rl=f.readlines()
    c=0
    for i in rl:
        c+=1
        if c%2!=0:
            print(i)
odd_lines()
```

Write a python program to read line by line from a given files file1 & file2 and write into file3. 

```python
f1=open("friends.txt","r")
data1=f1.readlines()
f1.close()

f2=open("friend.txt","r")
data2=f2.readlines()
f2.close()

f3=open("final.txt","w")
for i in range(len(data1)):
    f3.write(data1[i].strip()+"\n")
    f3.write(data2[i].strip()+"\n")
f3.close()


```
